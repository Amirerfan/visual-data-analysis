```python
import seaborn as sb
import pandas as pd
```

```python
sb.set_theme()
```

```python
solar_panel_data = pd.read_csv('./solar_panel.csv')
```

```python
# draw histogram for each column in the dataset using seaborn with size 20x20
for col in solar_panel_data.columns:
    sb.displot(solar_panel_data[col], kde=True, height=2, aspect=2)
```

```python
# draw histogram for each column in the dataset
solar_panel_data.hist(figsize=(20, 20))
```

```python
generated_power_grouped_by_date = solar_panel_data.groupby(['Day of Year', 'Year', 'Month', 'Day'])['Power Generated'].sum().reset_index()
generated_power_grouped_by_date.plot(x='Day of Year', y='Power Generated', figsize=(10, 10))
```

```python
maximum_generated_power_in_day = generated_power_grouped_by_date['Power Generated'].max()
generated_power_grouped_by_date[generated_power_grouped_by_date['Power Generated'] == maximum_generated_power_in_day]
```
